```python
#importing the required packages

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the data from the file and storing it as df
df = pd.read_csv("auto-dataset.csv")
```

```python
# Displaying the datainfo

df.info()
```

```python
import seaborn as sns
```

```python
#plotting all the variables against each other

sns.pairplot(df, kind = 'scatter')
```

```python
#importing Numpy to calculate the correlation

import numpy as np
```

```python
#correlation between cylinders and displacement variables

np.corrcoef(df['cylinders'], df['displacement'])
```

```python
#correlation between origin and year variables

np.corrcoef(df['origin'], df['year'])
```

```python
# importing LinearRegression package from the sklearn library

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# this is performing simple linear regression between mpg(target) and cylinders(feature)

#storing the valiues of the cylinders variable in x
x = df["cylinders"]

# storing the values of the mpg variable in y
y = df["mpg"]

# splitting the mpg values to x_train for training data and x_test for testing data (in half)

x_train = x[0:196]
x_test = x[197:392]

# splitting the cylinders values to y_train for training data and y_test for testing data (in half)

y_train = y[0:196]
y_test = y[197:392]

# we then create a linear regression object 

cylinders_predic_mpg_regre = LinearRegression()

# we train the model using the training sets
cylinders_predic_mpg_regre.fit(x_train.values.reshape(-1,1), y_train)

# we Make predictions using the testing set of cylinders
y_prediction = cylinders_predic_mpg_regre.predict(x_test.values.reshape(-1,1))


# we compute the correlation to measaure the relationship between the predictors and the outcome
print(np.corrcoef(df['cylinders'], df['mpg']))

# The R2 score of the model
print("R2 Score: %.2f" % r2_score(y_test, y_prediction))

# finally we plot the outputs
plt.scatter(x_test, y_test, color="red")
plt.plot(x_test, y_prediction, color="green", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# Multiple linear regression

# Definig the predictors
x = df[['cylinders', 'displacement','horsepower','weight','acceleration','year','origin']]

#splitting into training and testing sets
x_train = x[0:196]
x_test = x[197:392]

# Defining the response
y = df['mpg']

#splitting to test and  training sets

y_train = y[0:196]
y_test = y[197:392]

#we create a linear regression object
regr = LinearRegression()

#we train it with the training predictors and response
regr.fit(x_train, y_train)

# we Make predictions using the testing predictors
y_prediction = regr.predict(x_test)


# the R^2 score of the model
print("The R^2 Score of the model is :",regr.score(x_test, y_test))



regr.get_params(True)
```

```python
# manually go over test data and calculate residual
i = 0
for y_val in y_test:
    plt.scatter(y_val, y_val - y_prediction[i], color="blue") 
    i += 1
plt.xlabel("Fitted data")
plt.ylabel("Residual")
plt.show()
```
